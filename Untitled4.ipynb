Aqui inicializo o projeto com biblioteca python para a consulta da api do gitHub sobre dados do usuário.

```python
import requests

from google.colab import userdata

api_key = userdata.get('SECRET_KEY')
```

```python
# Solicitando o nome de usuário ao usuário
username = input("Digite seu usuário do GitHub: ")
```

```python
# URL da API do GitHub para obter informações do usuário
url = f'https://api.github.com/users/{username}'
url1 = f'https://api.github.com/users/{username}/repos'
# Cabeçalhos da requisição (recomendado pela documentação)
headers = {'Accept': 'application/vnd.github+json'}

# Realizando a requisição GET
response = requests.get(url, headers=headers)
response2 = requests.get(url1, headers=headers)

# Verificando o status da resposta
if response.status_code == 200:
  # Convertendo a resposta JSON para um dicionário Python

  #Aqui ficam armazenado os dados de
  user_data = response.json()
  user_repo = response2.json()
  # Imprimindo algumas informações do usuário
  print(user_repo)


  # Exemplo: acessando a URL da API para repositórios do usuário
  repos_url = user_data['repos_url']
  print(f"URL da API para Repositórios: {repos_url}")

else:
  print(f"Erro na requisição: {response.status_code}")
```

```python
# Solicitando Uma Autodescrição do Aluno
capacibilities = input("Forneça o maximo de informações sobre você, o que gosta, e quais suas areas de estudos favoritas:")
```

```python
# Import the Python SDK
import google.generativeai as genai
# Used to securely store your API key
#from google.colab import userdata

# code commented for studies fetures GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')

genai.configure(api_key=api_key)
```

Liste os modelos disponiveis de Ia para API

```python
for m in genai.list_models():
  if 'generateContent' in m.supported_generation_methods:
    print(m.name)
```

Aqui configuro Como desejo o retorno da resposta do google criando as variaveis de configuração

```python
generation_config ={
    "candidate_count": 1,
    "temperature": 0.5,
}
```

```python
safety_settings = {
    "HARASSMENT": "BLOCK_NONE",
    "HATE": "BLOCK_NONE",
    "SEXUAL": "BLOCK_NONE",
    "DANGEROUS": "BLOCK_NONE"
}
```

Aqui inicio qual modelo de inteligencia artificial e suas configurações

```python
model = genai.GenerativeModel(model_name="gemini-1.5-pro-latest",
                             generation_config=generation_config,
                             safety_settings=safety_settings)
```

Aqui incia o chat

```python
chat = model.start_chat(history=[])
```

```python
response = model.generate_content(f"Crie um cronograma de estudos baseados nas informações git é https://github.com/{username} e que" + str(user_data) + "são dados da api user do git, e que eu" + str(capacibilities) + f"e as informações dos repositórios são {user_repo} adicione links e sugestões conforme minhas a análise do git" )
print(response.text)
```

Aqui é inserida novas informações se a resposta não lhe agradar

```python
prompt = input("Qual sua duvida: ")
while prompt != "fim":
  response = chat.send_message(prompt)
  print("Resposta: ", response.text, "\n")
  prompt = input("Qual sua duvida: ")
```
